## Apply some cleaning to data to remove bad fits/unfolding

```python
n_samples = ubuf_ml_fit_mu_val[0].shape
sample_is_good = np.ones( shape=n_samples, dtype=bool )
```

```python


ub_val_mu = np.zeros( n_smearings )
ub_err_mu = np.zeros( n_smearings )
ub_rms_mu = np.zeros( n_smearings )
ub_val_sig2 = np.zeros( n_smearings )
ub_err_sig2 = np.zeros( n_smearings )
ub_rms_sig2 = np.zeros( n_smearings )

ub_diag_val_mu = np.zeros( n_smearings )
ub_diag_err_mu = np.zeros( n_smearings )
ub_diag_rms_mu = np.zeros( n_smearings )
ub_diag_val_sig2 = np.zeros( n_smearings )
ub_diag_err_sig2 = np.zeros( n_smearings )
ub_diag_rms_sig2 = np.zeros( n_smearings )

ub_ml_val_mu = np.zeros( n_smearings )
ub_ml_err_mu = np.zeros( n_smearings )
ub_ml_rms_mu = np.zeros( n_smearings )
ub_ml_val_sig2 = np.zeros( n_smearings )
ub_ml_err_sig2 = np.zeros( n_smearings )
ub_ml_rms_sig2 = np.zeros( n_smearings )



nn_ub_val_mu = np.zeros( n_smearings )
nn_ub_err_mu = np.zeros( n_smearings )
nn_ub_rms_mu = np.zeros( n_smearings )
nn_ub_val_sig2 = np.zeros( n_smearings )
nn_ub_err_sig2 = np.zeros( n_smearings )
nn_ub_rms_sig2 = np.zeros( n_smearings )

nn_ub_diag_val_mu = np.zeros( n_smearings )
nn_ub_diag_err_mu = np.zeros( n_smearings )
nn_ub_diag_rms_mu = np.zeros( n_smearings )
nn_ub_diag_val_sig2 = np.zeros( n_smearings )
nn_ub_diag_err_sig2 = np.zeros( n_smearings )
nn_ub_diag_rms_sig2 = np.zeros( n_smearings )

nn_ub_ml_val_mu = np.zeros( n_smearings )
nn_ub_ml_err_mu = np.zeros( n_smearings )
nn_ub_ml_rms_mu = np.zeros( n_smearings )
nn_ub_ml_val_sig2 = np.zeros( n_smearings )
nn_ub_ml_err_sig2 = np.zeros( n_smearings )
nn_ub_ml_rms_sig2 = np.zeros( n_smearings )






for si in range( n_smearings ) :
    
    
   #----

    rms_mu_ubuf = np.sqrt( np.cov(ubuf_chi2_fit_mu_val[si][sample_is_good[si]]) )
    rms_sig2_ubuf = np.sqrt( np.cov(ubuf_chi2_fit_sig2_val[si][sample_is_good[si]]) )

    ave_fit_err_mu_ubuf   = np.mean( ubuf_chi2_fit_mu_err[si][sample_is_good[si]] )
    ave_fit_err_sig2_ubuf = np.mean( ubuf_chi2_fit_sig2_err[si][sample_is_good[si]] )

    mean_mu_ubuf = np.mean( ubuf_chi2_fit_mu_val[si][sample_is_good[si]] )
    mean_sig2_ubuf = np.mean( ubuf_chi2_fit_sig2_val[si][sample_is_good[si]] )
    
   

    nn_rms_mu_ubuf = np.sqrt( np.cov(nn_ubuf_chi2_fit_mu_val[si][sample_is_good[si]]) )
    nn_rms_sig2_ubuf = np.sqrt( np.cov(nn_ubuf_chi2_fit_sig2_val[si][sample_is_good[si]]) )

    nn_ave_fit_err_mu_ubuf   = np.mean( nn_ubuf_chi2_fit_mu_err[si][sample_is_good[si]] )
    nn_ave_fit_err_sig2_ubuf = np.mean( nn_ubuf_chi2_fit_sig2_err[si][sample_is_good[si]] )

    nn_mean_mu_ubuf = np.mean( nn_ubuf_chi2_fit_mu_val[si][sample_is_good[si]] )
    nn_mean_sig2_ubuf = np.mean( nn_ubuf_chi2_fit_sig2_val[si][sample_is_good[si]] )







   #----

    rms_mu_ubuf_diag = np.sqrt( np.cov(ubuf_chi2_fit_diag_mu_val[si][sample_is_good[si]]) )
    rms_sig2_ubuf_diag = np.sqrt( np.cov(ubuf_chi2_fit_diag_sig2_val[si][sample_is_good[si]]) )

    ave_fit_err_mu_ubuf_diag   = np.mean( ubuf_chi2_fit_diag_mu_err[si][sample_is_good[si]] )
    ave_fit_err_sig2_ubuf_diag = np.mean( ubuf_chi2_fit_diag_sig2_err[si][sample_is_good[si]] )

    mean_mu_ubuf_diag = np.mean( ubuf_chi2_fit_diag_mu_val[si][sample_is_good[si]] )
    mean_sig2_ubuf_diag = np.mean( ubuf_chi2_fit_diag_sig2_val[si][sample_is_good[si]] )
    
    
    
    nn_rms_mu_ubuf_diag = np.sqrt( np.cov(nn_ubuf_chi2_fit_diag_mu_val[si][sample_is_good[si]]) )
    nn_rms_sig2_ubuf_diag = np.sqrt( np.cov(nn_ubuf_chi2_fit_diag_sig2_val[si][sample_is_good[si]]) )

    nn_ave_fit_err_mu_ubuf_diag   = np.mean( nn_ubuf_chi2_fit_diag_mu_err[si][sample_is_good[si]] )
    nn_ave_fit_err_sig2_ubuf_diag = np.mean( nn_ubuf_chi2_fit_diag_sig2_err[si][sample_is_good[si]] )

    nn_mean_mu_ubuf_diag = np.mean( nn_ubuf_chi2_fit_diag_mu_val[si][sample_is_good[si]] )
    nn_mean_sig2_ubuf_diag = np.mean( nn_ubuf_chi2_fit_diag_sig2_val[si][sample_is_good[si]] )

    
    
    
    
    
   #----
    
    rms_mu_ubuf_ml = np.sqrt( np.cov(ubuf_ml_fit_mu_val[si][sample_is_good[si]]) )
    rms_sig2_ubuf_ml = np.sqrt( np.cov(ubuf_ml_fit_sig2_val[si][sample_is_good[si]]) )

    ave_fit_err_mu_ubuf_ml   = np.mean( ubuf_ml_fit_mu_err[si][sample_is_good[si]] )
    ave_fit_err_sig2_ubuf_ml = np.mean( ubuf_ml_fit_sig2_err[si][sample_is_good[si]] )    

    mean_mu_ubuf_ml = np.mean( ubuf_ml_fit_mu_val[si][sample_is_good[si]] )
    mean_sig2_ubuf_ml = np.mean( ubuf_ml_fit_sig2_val[si][sample_is_good[si]] )


    
    nn_rms_mu_ubuf_ml = np.sqrt( np.cov(nn_ubuf_ml_fit_mu_val[si][sample_is_good[si]]) )
    nn_rms_sig2_ubuf_ml = np.sqrt( np.cov(nn_ubuf_ml_fit_sig2_val[si][sample_is_good[si]]) )

    nn_ave_fit_err_mu_ubuf_ml   = np.mean( nn_ubuf_ml_fit_mu_err[si][sample_is_good[si]] )
    nn_ave_fit_err_sig2_ubuf_ml = np.mean( nn_ubuf_ml_fit_sig2_err[si][sample_is_good[si]] )    

    nn_mean_mu_ubuf_ml = np.mean( nn_ubuf_ml_fit_mu_val[si][sample_is_good[si]] )
    nn_mean_sig2_ubuf_ml = np.mean( nn_ubuf_ml_fit_sig2_val[si][sample_is_good[si]] )

   
    
    

   #----
    
    ub_val_mu[si] = mean_mu_ubuf
    ub_err_mu[si] = ave_fit_err_mu_ubuf
    ub_rms_mu[si] = rms_mu_ubuf
    
    ub_val_sig2[si] = mean_sig2_ubuf
    ub_err_sig2[si] = ave_fit_err_sig2_ubuf
    ub_rms_sig2[si] = rms_sig2_ubuf    


    
    nn_ub_val_mu[si] = nn_mean_mu_ubuf
    nn_ub_err_mu[si] = nn_ave_fit_err_mu_ubuf
    nn_ub_rms_mu[si] = nn_rms_mu_ubuf
    
    nn_ub_val_sig2[si] = nn_mean_sig2_ubuf
    nn_ub_err_sig2[si] = nn_ave_fit_err_sig2_ubuf
    nn_ub_rms_sig2[si] = nn_rms_sig2_ubuf    
    
    
    
    
    
    
    
    
   #----
    
    ub_diag_val_mu[si] = mean_mu_ubuf_diag
    ub_diag_err_mu[si] = ave_fit_err_mu_ubuf_diag
    ub_diag_rms_mu[si] = rms_mu_ubuf_diag
    
    ub_diag_val_sig2[si] = mean_sig2_ubuf_diag
    ub_diag_err_sig2[si] = ave_fit_err_sig2_ubuf_diag
    ub_diag_rms_sig2[si] = rms_sig2_ubuf_diag 
    
    
    nn_ub_diag_val_mu[si] = nn_mean_mu_ubuf_diag
    nn_ub_diag_err_mu[si] = nn_ave_fit_err_mu_ubuf_diag
    nn_ub_diag_rms_mu[si] = nn_rms_mu_ubuf_diag
    
    nn_ub_diag_val_sig2[si] = nn_mean_sig2_ubuf_diag
    nn_ub_diag_err_sig2[si] = nn_ave_fit_err_sig2_ubuf_diag
    nn_ub_diag_rms_sig2[si] = nn_rms_sig2_ubuf_diag 
    

    
    
    
    
   #----
    
    ub_ml_val_mu[si] = mean_mu_ubuf_ml
    ub_ml_err_mu[si] = ave_fit_err_mu_ubuf_ml
    ub_ml_rms_mu[si] = rms_mu_ubuf_ml
    
    ub_ml_val_sig2[si] = mean_sig2_ubuf_ml
    ub_ml_err_sig2[si] = ave_fit_err_sig2_ubuf_ml
    ub_ml_rms_sig2[si] = rms_sig2_ubuf_ml


    nn_ub_ml_val_mu[si] = nn_mean_mu_ubuf_ml
    nn_ub_ml_err_mu[si] = nn_ave_fit_err_mu_ubuf_ml
    nn_ub_ml_rms_mu[si] = nn_rms_mu_ubuf_ml
    
    nn_ub_ml_val_sig2[si] = nn_mean_sig2_ubuf_ml
    nn_ub_ml_err_sig2[si] = nn_ave_fit_err_sig2_ubuf_ml
    nn_ub_ml_rms_sig2[si] = nn_rms_sig2_ubuf_ml

    
    



    print('\n\n')
    
    print(' =========== Smearing %6.4f' % smearing_vals[si])

    print('     Unbinned unfolding,         mu value mean %8.4f, RMS: %8.4f, ave fit err %6.4f' % 
          (mean_mu_ubuf, rms_mu_ubuf, ave_fit_err_mu_ubuf) )
    print('  NN Unbinned unfolding,         mu value mean %8.4f, RMS: %8.4f, ave fit err %6.4f' % 
          (nn_mean_mu_ubuf, nn_rms_mu_ubuf, nn_ave_fit_err_mu_ubuf) )

    
    
    print('     Unbinned unfolding, diag.   mu value mean %8.4f, RMS: %8.4f, ave fit err %6.4f' % 
          (mean_mu_ubuf_diag, rms_mu_ubuf_diag, ave_fit_err_mu_ubuf_diag) )    
    print('  NN Unbinned unfolding, diag.   mu value mean %8.4f, RMS: %8.4f, ave fit err %6.4f' % 
          (nn_mean_mu_ubuf_diag, nn_rms_mu_ubuf_diag, nn_ave_fit_err_mu_ubuf_diag) )
    
    
    
    print('     Unbinned unf. ML fit,       mu value mean %8.4f, RMS: %8.4f, ave fit err %6.4f' % 
          (mean_mu_ubuf_ml, rms_mu_ubuf_ml, ave_fit_err_mu_ubuf_ml) )
    print('  NN Unbinned unf. ML fit,       mu value mean %8.4f, RMS: %8.4f, ave fit err %6.4f' % 
          (nn_mean_mu_ubuf_ml, nn_rms_mu_ubuf_ml, nn_ave_fit_err_mu_ubuf_ml) )
    
    
    
    
    print('\n')
    print('     Unbinned unfolding,       sig2 value mean %8.4f, RMS: %8.4f, ave fit err %6.4f' % 
          ( mean_sig2_ubuf, rms_sig2_ubuf, ave_fit_err_sig2_ubuf) )
    print('  NN Unbinned unfolding,       sig2 value mean %8.4f, RMS: %8.4f, ave fit err %6.4f' % 
          ( nn_mean_sig2_ubuf, nn_rms_sig2_ubuf, nn_ave_fit_err_sig2_ubuf) )
    
    
    
    print('     Unbinned unfolding, diag. sig2 value mean %8.4f, RMS: %8.4f, ave fit err %6.4f' % 
          ( mean_sig2_ubuf_diag, rms_sig2_ubuf_diag, ave_fit_err_sig2_ubuf_diag) )
    print('  NN Unbinned unfolding, diag. sig2 value mean %8.4f, RMS: %8.4f, ave fit err %6.4f' % 
          ( nn_mean_sig2_ubuf_diag, nn_rms_sig2_ubuf_diag, nn_ave_fit_err_sig2_ubuf_diag) )
    
    
    
    
    print('     Unbinned unf. ML  ,       sig2 value mean %8.4f, RMS: %8.4f, ave fit err %6.4f' % 
          ( mean_sig2_ubuf_ml, rms_sig2_ubuf_ml, ave_fit_err_sig2_ubuf_ml) )
    print('  NN Unbinned unf. ML  ,       sig2 value mean %8.4f, RMS: %8.4f, ave fit err %6.4f' % 
          ( nn_mean_sig2_ubuf_ml, nn_rms_sig2_ubuf_ml, nn_ave_fit_err_sig2_ubuf_ml) )
  




print('\n\n')


```
